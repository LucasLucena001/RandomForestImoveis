```python
import pandas as pd
import numpy as np
```

```python
casa = pd.read_excel('/content/house (1).xlsx')
casa.head()
```

```python
casa.info()
```

```python
casa.describe()
```

```python
casa = casa.dropna()
```

```python
casa = pd.get_dummies(casa)
casa.head()
```

```python
# prompt: transfromar essas colunas de boolean para int ExterQual_Ex	ExterQual_Fa	ExterQual_Gd	ExterQual_TA	ExterCond_Ex	ExterCond_Fa	ExterCond_Gd	ExterCond_TA

casa[['ExterQual_Ex', 'ExterQual_Fa', 'ExterQual_Gd', 'ExterQual_TA', 'ExterCond_Ex', 'ExterCond_Fa', 'ExterCond_Gd', 'ExterCond_TA']] = casa[['ExterQual_Ex', 'ExterQual_Fa', 'ExterQual_Gd', 'ExterQual_TA', 'ExterCond_Ex', 'ExterCond_Fa', 'ExterCond_Gd', 'ExterCond_TA']].astype(int)
```

```python
labels= np.array(casa['SalePrice'])

features = casa.drop('SalePrice', axis = 1)

feature_list = list(casa.columns)

casa = np.array(features)
```

```python
from sklearn.model_selection import train_test_split


train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state =42)
```

```python
from sklearn.datasets import make_moons
from sklearn import metrics
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
rf = RandomForestRegressor(n_estimators=1000, random_state =42)
rf.fit(train_features, train_labels)
```

```python
prediction_rf = rf.predict(test_features)

error_rf = abs(prediction_rf - test_labels)

r_sq = rf.score(features,labels)

print('R2:', r_sq)
print('MAE: ', metrics.mean_absolute_error(test_labels, prediction_rf))
print('MSE: ', metrics.mean_squared_error(test_labels, prediction_rf))
```

```python
ada = AdaBoostRegressor(n_estimators = 1000)
ada.fit(train_features, train_labels)

ada_pred = ada.predict(test_features)
```

```python
error_ada = abs(ada_pred - test_labels)

r_sq_ada = ada.score(features,labels)

print('R2:', r_sq_ada)
print('MAE: ', metrics.mean_absolute_error(test_labels, ada_pred))
print('MSE: ', metrics.mean_squared_error(test_labels, ada_pred))
```

```python
gbr = GradientBoostingRegressor(n_estimators = 1000)
gbr.fit(train_features, train_labels)
gbr_pred = gbr.predict(test_features)
```
